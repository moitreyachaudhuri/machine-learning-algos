```python
#loading the data into the data frame
import pandas as pd
data=pd.read_csv("A2Data_train.csv")
data.head()
```

```python
data.shape
```

```python
#selecting the appropriate rows and columns for feature matrix and label vector
features= data.loc[ : , data.columns != '1.9741']
features.shape
```

```python
labels=data.loc[ : , data.columns== '1.9741']
labels.shape
```

```python
#converting our features and labels to numpy array for matrix computations
import numpy as np
x=np.array(features)
y=np.array(labels)
```

```python
#using the closed form expression to compute the coefficient vector w_ml
X=x.T
w_ml=(np.linalg.inv(X@x))@X@y
w_ml
```

```python
#error in our case
((x@w_ml-y).T)@(x@w_ml-y)
```

```python
#just visualising the individual error components
x@w_ml-y
```
